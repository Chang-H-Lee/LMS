### 2. Input Normalization

```python
# 학습 데이터와 검증 데이터에 normalizing 과 resizing 적용
def normalize_and_resize_img(image, label):
    """Normalizes images: `uint8` -> `float32`."""
    image = tf.image.resize(image, (224,224))
    image = tf.cast(image, tf.float32) / 255.
    return image, label
```

```python
# 전처리 포함한 배치 구성
def apply_normalize_on_dataset(ds, is_test=False, batch_size=16):
    ds = ds.map(
        normalize_and_resize_img, 
        num_parallel_calls=1
    )
    ds = ds.batch(batch_size)
    if not is_test:
        ds = ds.repeat()
        ds = ds.shuffle(200)
    ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
    return ds
```

```python
# 데이터셋에 전처리와 배치 처리 적용
BATCH_SIZE = 32
ds_train = apply_normalize_on_dataset(ds_train, batch_size=BATCH_SIZE)
ds_test = apply_normalize_on_dataset(ds_test, batch_size=BATCH_SIZE)
```

```python
# 구성된 배치 모양 확인
for input in ds_train.take(1):
    image, label = input
    print(image.shape)
    print(label.shape)
```

### 3. ResNet-34 Plain 구성

```python
# 하이퍼 파라메터
BATCH_SIZE = 32  # 256
EPOCHS = 10  # 15
LR = 0.1
OPT_DECAY = 0.0001
MOMENTUM = 0.9
L2_WEIGHT_DECAY = 1e-4
BATCH_NORM_DECAY = 0.9
BATCH_NORM_EPSILON = 1e-5

ACTIVATION = 'sigmoid'
NUM_CLASSES = 1
```

ResNet 레이어 구성

image.png

[ 참 고 ]    
ResNet-34와 ResNet-50은 모두 conv block이라고 불리는 블록 구조를 각각 3, 4, 6, 3개씩 반복해서 쌓은 형태라는 공통점을 가지고 있습니다. 그러나 ResNet-34와 ResNet-50의 블록 내부 구성은 약간 다릅니다. ResNet-34의 경우 Block은 3x3 kernel인 Convolution layer로만 구성되어있지만, ResNet-50은 1x1 Convolution이 앞뒤로 붙어 더 많은 레이어를 한 블록 내에 가지게 됩니다

#### ResNet-34 Layers 구성

image.png

```python
# Conv 블록 생성 함수
def conv_block(input_layer, 
                channel, 
                kernel_size, 
                strides=1, 
                activation='relu',
                l2_weight_decay=1e-4,
                batch_norm_decay=0.9,
                batch_norm_epsilon=1e-5):
    
    x = keras.layers.Conv2D(
        filters=channel,
        kernel_size=kernel_size,
        kernel_initializer='he_normal',
        kernel_regularizer=keras.regularizers.l2(l2_weight_decay),
        padding='same',
        strides=strides,
        use_bias=False,
    )(input_layer)
    
    x = keras.layers.BatchNormalization(
        # momentum=batch_norm_decay,
        # epsilon=batch_norm_epsilon,
    )(x)
    
    if activation:
        x = keras.layers.Activation(activation)(x)
    
    return x
```

```python
# 블록 구성 (ResNet-34, Plain)

def build_net_block_34P(input_layer, 
                        cnn_count=3, 
                        channel=64, 
                        block_num=0, 
                                ):
    x = input_layer

    # 첫 번째 conv_layer 에서 strides=2 설정해서 너비와 높이를 줄임
    # => 필터의 개수(=채널)가 증가할 때마다 너비와 높이를 줄임

    for i in range(cnn_count):
        if block_num > 0 and i == 0:
            x = conv_block(x, channel, (3,3), strides=2)
            x = conv_block(x, channel, (3,3))
        else:
            x = conv_block(x, channel, (3,3))
            x = conv_block(x, channel, (3,3))

    return x
```

```python
# 모델 생성 함수

def build_net_34P(input_shape=(32,32,3),
                cnn_count_list=[3,4,6,3],
                channel_list=[64,128,256,512],
                num_classes=1, 
                activation='sigmoid',
                name='ResNet_34'):
    
    # 모델 생성 전, config list들이 같은 길이인지 확인
    assert len(cnn_count_list) == len(channel_list)
    
    # input layer 생성
    input_layer = keras.layers.Input(shape=input_shape)
    
    # first layer 설정
    x = conv_block(input_layer, 64, (7,7), strides=2)
    x = keras.layers.MaxPool2D(pool_size=(2,2), strides=2)(x)
    
    # config list들의 길이만큼 반복
    for block_num, (cnn_count, channel) in enumerate(zip(cnn_count_list, channel_list)):
        x = build_net_block_34P(x, 
                                cnn_count=cnn_count, 
                                channel=channel, 
                                block_num=block_num)
   
    x = keras.layers.GlobalAveragePooling2D()(x)
    x = keras.layers.Dense(num_classes, 
                        activation=activation, 
                        kernel_initializer='he_normal'
                        )(x)
    
    model = keras.Model(inputs=input_layer, outputs=x, name=name)
    
    return model
```

```python
# 모델 생성 및 확인
resnet_34P = build_net_34P(
    input_shape=(224,224,3),
    num_classes=NUM_CLASSES,
    activation=ACTIVATION,
    name='ResNet_34P',
)

resnet_34P.summary()
```

### 4. ResNet-34 구성

```python
# 블록 구성 (ResNet-34)

def build_net_block_34(input_layer, 
                       cnn_count=3, 
                       channel=64, 
                       block_num=0, 
                                ):
    x = input_layer
    
    for i in range(cnn_count):
        if block_num > 0 and i == 0:
            shortcut = conv_block(x, channel, (1,1), strides=2, activation=None)
            x = conv_block(x, channel, (3,3), strides=2)
            x = conv_block(x, channel, (3,3), activation=None)
        else:
            shortcut = x
            x = conv_block(x, channel, (3,3))
            x = conv_block(x, channel, (3,3), activation=None)
        
        x = keras.layers.Add()([x, shortcut])
        x = keras.layers.Activation('relu')(x)

    return x
```

```python
# 모델 생성 함수

def build_net_34(input_shape=(32,32,3),
                cnn_count_list=[3,4,6,3],
                channel_list=[64,128,256,512],
                num_classes=1, 
                activation='sigmoid',
                name='ResNet_34'):
    
    # 모델 생성 전, config list들이 같은 길이인지 확인
    assert len(cnn_count_list) == len(channel_list)
    
    # input layer 생성
    input_layer = keras.layers.Input(shape=input_shape)
    
    # first layer 설정
    x = conv_block(input_layer, 64, (7,7), strides=2)
    x = keras.layers.MaxPool2D(pool_size=(2,2), strides=2)(x)
    
    # config list들의 길이만큼 반복
    for block_num, (cnn_count, channel) in enumerate(zip(cnn_count_list, channel_list)):
        x = build_net_block_34(x, 
                               cnn_count=cnn_count, 
                               channel=channel, 
                               block_num=block_num)
   
    x = keras.layers.GlobalAveragePooling2D()(x)
    x = keras.layers.Dense(num_classes, 
                        activation=activation, 
                        kernel_initializer='he_normal'
                        )(x)
    
    model = keras.Model(inputs=input_layer, outputs=x, name=name)
    
    return model
```

```python
# 모델 생성 및 확인
resnet_34 = build_net_34(
    input_shape=(224,224,3),
    num_classes=NUM_CLASSES,
    activation=ACTIVATION,
    name='ResNet_34',
)

resnet_34.summary()
```

### 5. ResNet-50 Plain 구성

```python
# 블록 구성 (ResNet-50, Plain)

def build_net_block_50P(input_layer, 
                        cnn_count=3, 
                        channel=64, 
                        block_num=0, 
                                ):
    x = input_layer

    for i in range(cnn_count):
        if block_num > 0 and i == 0:
            x = conv_block(x, channel, (1,1), strides=2)
            x = conv_block(x, channel, (3,3))
            x = conv_block(x, channel*4, (3,3))
        else:
            x = conv_block(x, channel, (1,1))
            x = conv_block(x, channel, (3,3))
            x = conv_block(x, channel*4, (1,1))

    return x
```

```python
# 모델 생성 함수

def build_net_50P(input_shape=(32,32,3),
                cnn_count_list=[3,4,6,3],
                channel_list=[64,128,256,512],
                num_classes=1, 
                activation='sigmoid',
                name='ResNet_50'):
    
    # 모델 생성 전, config list들이 같은 길이인지 확인
    assert len(cnn_count_list) == len(channel_list)
    
    # input layer 생성
    input_layer = keras.layers.Input(shape=input_shape)
    
    # first layer 설정
    x = conv_block(input_layer, 64, (7,7), strides=2)
    x = keras.layers.MaxPool2D(pool_size=(2,2), strides=2)(x)
    
    # config list들의 길이만큼 반복
    for block_num, (cnn_count, channel) in enumerate(zip(cnn_count_list, channel_list)):
        x = build_net_block_50P(x, 
                                cnn_count=cnn_count, 
                                channel=channel, 
                                block_num=block_num)
   
    x = keras.layers.GlobalAveragePooling2D()(x)
    x = keras.layers.Dense(num_classes, 
                        activation=activation, 
                        kernel_initializer='he_normal'
                        )(x)
    
    model = keras.Model(inputs=input_layer, outputs=x, name=name)
    
    return model
```

```python
# 모델 생성 및 확인
resnet_50P = build_net_50P(
    input_shape=(224,224,3),
    num_classes=NUM_CLASSES,
    activation=ACTIVATION,
    name='ResNet_50P',
)

resnet_50P.summary()
```

### 6. ResNet-50 구성

```python
# 블록 구성 (ResNet-50)

def build_net_block_50(input_layer, 
                       cnn_count=3, 
                       channel=64, 
                       block_num=0, 
                                ):
    x = input_layer
    
    for i in range(cnn_count):
        if i == 0:
            shortcut = conv_block(x, channel*4, (1,1), strides=2, activation=None)
            x = conv_block(x, channel, (1,1), strides=2)
            x = conv_block(x, channel, (3,3))
            x = conv_block(x, channel*4, (1,1), activation=None)
        else:
            shortcut = x
            x = conv_block(x, channel, (1,1))
            x = conv_block(x, channel, (3,3))
            x = conv_block(x, channel*4, (1,1), activation=None)
        
        x = keras.layers.Add()([x, shortcut])
        x = keras.layers.Activation('relu')(x)

    return x
```

```python
# 모델 생성 함수

def build_net_50(input_shape=(32,32,3),
                cnn_count_list=[3,4,6,3],
                channel_list=[64,128,256,512],
                num_classes=1, 
                activation='sigmoid',
                name='ResNet_50'):
    
    # 모델 생성 전, config list들이 같은 길이인지 확인
    assert len(cnn_count_list) == len(channel_list)
    
    # input layer 생성
    input_layer = keras.layers.Input(shape=input_shape)
    
    # first layer 설정
    x = conv_block(input_layer, 64, (7,7), strides=2)
    x = keras.layers.MaxPool2D(pool_size=(2,2), strides=2)(x)
    
    # config list들의 길이만큼 반복
    for block_num, (cnn_count, channel) in enumerate(zip(cnn_count_list, channel_list)):
        x = build_net_block_50(x, 
                               cnn_count=cnn_count, 
                               channel=channel, 
                               block_num=block_num)
   
    x = keras.layers.GlobalAveragePooling2D()(x)
    x = keras.layers.Dense(num_classes, 
                        activation=activation, 
                        kernel_initializer='he_normal'
                        )(x)
    
    model = keras.Model(inputs=input_layer, outputs=x, name=name)
    
    return model
```

```python
# 모델 생성 및 확인
resnet_50 = build_net_50(
    input_shape=(224,224,3),
    num_classes=NUM_CLASSES,
    activation=ACTIVATION,
    name='ResNet_50',
)

resnet_50.summary()
```

### 7. Performance 비교

```python
# 학습 함수 구성
def compile_and_fit(model, ds_train, ds_test, ds_info, learning_rate=0.001, momentum= 0.9, opt_decay=0.0001, batch_size=128, epochs=15):
    model.compile(
        loss='binary_crossentropy',
        optimizer=tf.keras.optimizers.SGD(
            learning_rate=learning_rate,
            momentum=momentum,
            decay=opt_decay,
            clipnorm=1.
        ),
        metrics=['accuracy'],
    )

    history = model.fit(
        ds_train,
        steps_per_epoch=int(ds_info.splits['train[:80%]'].num_examples/batch_size),
        validation_steps=int(ds_info.splits['train[80%:]'].num_examples/batch_size),
        epochs=epochs,
        validation_data=ds_test,
        verbose=1,
        use_multiprocessing=True,
    )

    return history
```

### 8. 결론 및 회고
  * 논문(Deep Residual Learning for Image Recognition)에서 주장한 깊기만 한 layer의 Train 및 Test 성과가 좋지 않음을 확인(34-layer vs 50-layer)
  * skip connection을 활용한 Residual Network의 성과가 더 좋게 나옴을 확인(plain layer 대비)
  * ResNet-34와 ReNet-50의 비교는 ResNet-34의 성과가 더 좋게 나왔으나, 학습시간 제한으로 epoch을 10까지만 하고, 빠른 학습을 위해 learning rate를 0.1로 한 영향이 있었을 것으로 추정됨. learning rate를 줄이고, epoch을 늘리면 ReNet-50 성과가 좋을 것으로 기대됨(시간 될 때 돌려봐야겠음)
  * layer를 쌓아감에 있어서, For문과 If문을 적절히 사용하는 방법 및 Class를 활용하는 방법을 배워야겠음     
  * (dog and cat) dataset update로 다운로드 안되어, 호원님이 좋은 방법을 알려주어 도움을 받았음 ==> setattr(tfds.image_classification.cats_vs_dogs, '_URL',"https://download.microsoft.com/download/3/E/1/3E1C3F21-ECDB-4869-8368-6DEBA77B919F/kagglecatsanddogs_5340.zip") 로 tensorflow_datasets 의 주소만 바꿔주는 방법!!
