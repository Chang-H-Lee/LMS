# 1. 정면사진의 경우

## 1-1 얼굴이 포함된 사진 준비

```python
my_image_path = '/content/drive/MyDrive/LMS/E_08/camera_sticker/images/self_image_01_resize.png'
img_bgr = cv2.imread(my_image_path)    # OpenCV로 이미지를 불러옵니다
img_show = img_bgr.copy()      # 출력용 이미지를 따로 보관합니다
plt.imshow(img_bgr)
plt.show()
```

왜 얼굴색이 푸른빛일까? : opencv는 이미지 채널을 BGR 순으로 사용하기 때문

```python
# plt.imshow 이전에 RGB 이미지로 바꾸는 것을 잊지마세요. 
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

정상적인 사진표시를 위해 RGB형태로 변환

## 1-2 얼굴 검출 face detection

```python
# detector를 선언합니다
detector_hog = dlib.get_frontal_face_detector()
```

얼굴위치를 찾기 위해 dlib의 face detector 사용 : HOG(Histogram of Oriented Gradients)와 SVM(Support Vector Machine)을 사용한다고 함. HOG는 이미지에서 색상의 변화량을 나타낸 것으로 이미지로부터 물체의 특징을 잘 잡아냄. SVM은 선형 분류기로서 한 이미지를 다차원 공간의 한 벡터로 보고 얼굴이냐 아니냐를 구분

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   # (image, num of image pyramid)
```

detector_hog를 이용해 얼굴의 bounding box를 추출

```python
# 찾은 얼굴 영역 박스 리스트
# 여러 얼굴이 있을 수 있습니다
print(dlib_rects)   

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

찾은 얼굴을 화면에 출력: 우선은 1개 얼굴을 찾았음.   
dlib.rectangle객체는 left(), top(), right(), bottom(), height(), width() 등의 멤버 함수를 포함

## 1-3 얼굴 랜드마크 (face landmark)

```python
model_path = '/content/drive/MyDrive/LMS/E_08/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
```

저장한 landmark 모델을 불러옴: landmark_predictor 는 RGB 이미지와 dlib.rectangle을 입력 받아 dlib.full_object_detection 를 반환

```python
list_landmarks = []

# 얼굴 영역 박스 마다 face landmark를 찾아냅니다
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    # face landmark 좌표를 저장해둡니다
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))
```

list_points 는 tuple (x, y) 68개로 이루어진 리스트
face_dot.PNG

```python
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

랜드마크가 이미지의 각 위치에 표시됨

## 1-4 스티커 적용하기

```python
# 좌표 확인
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[30]) # 코의 index는 30 입니다
    # print (landmark[51]) # 윗입술 index는 51 입니다
    x = landmark[30][0]
    # x = landmark[51][0]
    y = landmark[30][1] + dlib_rect.height()//2    # 수염을 붙이기 위해 스티커 아래로 이동  
    # y = landmark[51][1] + dlib_rect.height()//2    
    w = h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
```

스티커 위치 : x = x(코), y = y(코)+(height)/2

스티커 크기 : width = height = box 넓이


```python
sticker_path = '/content/drive/MyDrive/LMS/E_08/camera_sticker/images/cat_whiskers.png'
img_sticker = cv2.imread(sticker_path) # 스티커 이미지를 불러옵니다
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)
```

준비한 스티커를 불러오고 크기를 박스크기로 조정

```python
plt.imshow(img_sticker)
```

```python
refined_x = x - w // 2      # left
refined_y = y - h           # top
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

원본이미지에 스티커 이미지 추가하기 위해 x,y좌표 조정(시작점이 top-left좌표)

```python
if refined_x < 0: 
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :]
    refined_y = 0

print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

사진화면을 벗어난 스티커 제거: 사진범위내로 확인됨

```python
# 길어서 복잡해 보이지만 img_show[from:to] 형식입니다
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
```

np.where 함수를 이용하여 원본그림에 스티커를 붙임

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

설정한 좌표 위치에 스티커가 위치하는지 확인

```python
sticker_area = img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    cv2.addWeighted(sticker_area,0.5,np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8),0.5,0)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

## 2-2 얼굴영역찾기 함수 구성

```python
def find_face(name, width, height):

    my_image_path =f'/content/drive/MyDrive/LMS/E_08/camera_sticker/images/scant_img/{name}.png'
    img_bgr = cv2.imread(my_image_path)    #- OpenCV로 이미지를 읽어서
    img_bgr = cv2.resize(img_bgr, (width, height))    # width x height 의 크기로 Resize
    img_show = img_bgr.copy()      #- 출력용 이미지 별도 보관
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  # BGR 값을 RGB로 변환

    detector_hog = dlib.get_frontal_face_detector()   #- detector 선언

    dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)

    for dlib_rect in dlib_rects:
        l = dlib_rect.left()
        t = dlib_rect.top()
        r = dlib_rect.right()
        b = dlib_rect.bottom()

        cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

    img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
    plt.imshow(img_show_rgb)
    plt.show()
```

## 2-3. 얼굴랜드마크 표시하기 함수 구성

```python
def find_landmark(name, width, height):

    my_image_path =f'/content/drive/MyDrive/LMS/E_08/camera_sticker/images/scant_img/{name}.png'
    img_bgr = cv2.imread(my_image_path)    #- OpenCV로 이미지를 읽어서
    img_bgr = cv2.resize(img_bgr, (width, height))    # width x height 의 크기로 Resize
    img_show = img_bgr.copy()      #- 출력용 이미지 별도 보관
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  # BGR 값을 RGB로 변환

    detector_hog = dlib.get_frontal_face_detector()   #- detector 선언

    dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)

    for dlib_rect in dlib_rects:
        l = dlib_rect.left()
        t = dlib_rect.top()
        r = dlib_rect.right()
        b = dlib_rect.bottom()

        cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)
 
    model_path = '/content/drive/MyDrive/LMS/E_08/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
    landmark_predictor = dlib.shape_predictor(model_path)         # 저장된 landmark 모델 불러오기
    
    list_landmarks = []
    for dlib_rect in dlib_rects:
        points = landmark_predictor(img_rgb, dlib_rect)
        list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        list_landmarks.append(list_points)

        for landmark in list_landmarks:
            for idx, point in enumerate(landmark):
                cv2.circle(img_show, point, 2, (0, 255, 255), -1) # yellow

        img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
        plt.imshow(img_show_rgb)
        plt.show()
```

```python
for name in scant_list:
    find_landmark(name,640,480)   # 640 x 480 크기의 이미지로 resize 해서 불러온 이미지 얼굴 영역 랜드마크 표시
```

## 2-4 스티커 붙이기 함수 구성

```python
def sticker(name, width, height):

    my_image_path =f'/content/drive/MyDrive/LMS/E_08/camera_sticker/images/scant_img/{name}.png'
    img_bgr = cv2.imread(my_image_path)    #- OpenCV로 이미지를 읽어서
    img_bgr = cv2.resize(img_bgr, (width, height))    # width x height 의 크기로 Resize
    img_show = img_bgr.copy()      #- 출력용 이미지 별도 보관
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  # BGR 값을 RGB로 변환

    detector_hog = dlib.get_frontal_face_detector()   #- detector 선언

    dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)

    for dlib_rect in dlib_rects:
        l = dlib_rect.left()
        t = dlib_rect.top()
        r = dlib_rect.right()
        b = dlib_rect.bottom()

        cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)
        
    model_path = '/content/drive/MyDrive/LMS/E_08/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
    landmark_predictor = dlib.shape_predictor(model_path)         # 저장된 landmark 모델 불러오기
    
    list_landmarks = []
    for dlib_rect in dlib_rects:
        points = landmark_predictor(img_rgb, dlib_rect)
        list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        list_landmarks.append(list_points)

        for landmark in list_landmarks:
            for idx, point in enumerate(landmark):
                cv2.circle(img_show, point, 2, (0, 255, 255), -1) # yellow

    for dlib_rect, landmark in zip(dlib_rects, list_landmarks):

        x = landmark[30][0]   # 코의 인덱스
        y = landmark[30][1] + dlib_rect.height()//2    # 수염을 붙이기 위해 스티커 아래로 이동  
    
        w = h = dlib_rect.width()
        
    img_sticker = cv2.imread(sticker_path)
    img_sticker = cv2.resize(img_sticker, (w,h))
    
    refined_x = x - w // 2      # left
    refined_y = y - h           # top
       
    sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
    img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
      
    plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
    plt.show()
```

## 2-5 회전함수 및 투명도 적용

```python
# 회전함수
def im_rotate(img, degree):
    h, w = img.shape[:-1]
    centerRotatePT = int(w / 2), int(h / 2)
    new_h, new_w = h, w 
    rotatefigure = cv2.getRotationMatrix2D(centerRotatePT, degree, 1) 
    result = cv2.warpAffine(img, rotatefigure, (new_w, new_h)) 
    return result
```

```python
# 투명도 적용 스티커 부착 최종함수
def sticker_refine(name, width, height):

    my_image_path =f'/content/drive/MyDrive/LMS/E_08/camera_sticker/images/scant_img/{name}.png'
    img_bgr = cv2.imread(my_image_path)    #- OpenCV로 이미지를 읽어서
    img_bgr = cv2.resize(img_bgr, (width, height))    # width x height 의 크기로 Resize
    img_show = img_bgr.copy()      #- 출력용 이미지 별도 보관
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  # BGR 값을 RGB로 변환

    detector_hog = dlib.get_frontal_face_detector()   #- detector 선언

    dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)

    for dlib_rect in dlib_rects:
        l = dlib_rect.left()
        t = dlib_rect.top()
        r = dlib_rect.right()
        b = dlib_rect.bottom()

        cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)
        
    model_path = '/content/drive/MyDrive/LMS/E_08/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
    landmark_predictor = dlib.shape_predictor(model_path)         # 저장된 landmark 모델 불러오기
    
    list_landmarks = []
    for dlib_rect in dlib_rects:
        points = landmark_predictor(img_rgb, dlib_rect)
        list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        list_landmarks.append(list_points)

        # for landmark in list_landmarks:
        #     for idx, point in enumerate(landmark):
        #         cv2.circle(img_show, point, 2, (0, 255, 255), -1) # yellow

    for dlib_rect, landmark in zip(dlib_rects, list_landmarks):

        x = landmark[30][0]
        y = landmark[30][1]
   
        w = h = dlib_rect.width()

        m1 = landmark[33][1]   # 인중
        m2 = landmark[51][1]   # 윗입술
        m3 = landmark[2]       # 우측귀
        m4 = landmark[14]      # 좌측귀
        
        img_sticker = cv2.imread(sticker_path)
        img_sticker = cv2.resize(img_sticker, (w,h))

        # 회전
        s=np.array(m4)-np.array(m3)
        se=np.arctan(s[1]/s[0])*(180/np.pi)
        img_sticker = np.where(img_sticker==255, 0, 255).astype(np.uint8) # 색반전됨 ㅠㅠ
        img_sticker = im_rotate(img_sticker,-se)
        img_sticker = np.where(img_sticker==255, 0, 255).astype(np.uint8) # 다시 색반전

        refined_x = x - w // 2    # left
        refined_y = (m1 + m2 - h) // 2    # top
        
        sticker_area = img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
        img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
        cv2.addWeighted(sticker_area,0.5,np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8),0.5,0)
      
    plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
for name in scant_list:
    sticker_refine(name,640,480)
```
